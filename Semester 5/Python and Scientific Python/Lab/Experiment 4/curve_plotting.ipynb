```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y=range(2,22,2)
y2=range(10,20,1)

plt.plot(y)
plt.plot(y2, linewidth=5)
plt.legend(["y", "y2"])
plt.show()
```

```python
plt.plot(y)
plt.plot(y2, linewidth=5)
plt.legend(["y", "y2"], loc="lower right")
plt.show()
```

```python
plt.plot(y)
plt.plot(y2, linewidth=5)
plt.legend(["y", "y2"], bbox_to_anchor=(0.6, 1))
plt.show()
```

```python
def f1(t):
    return t**2*np.exp(-t**2)

def f2(t):
    return t**2*f1(t)

t = np.linspace(0,3,51)
y1=f1(t)
y2=f2(t)

plt.plot(t, y1, "r-")
plt.plot(t, y2, "bo")
plt.xlabel("t")
plt.ylabel("y")
plt.legend(["t^2*exp(-t^2)", "t^4*exp(-t^2)"])
plt.title("Plotting two curves in the same plot")
plt.show()
```
